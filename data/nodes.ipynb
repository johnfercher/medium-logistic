## Imports

```python
import pandas as pd
```

## Load Data

```python
df = pd.read_csv('nodes.csv')
df
```

## Normalize Population

```python
normalized = df
normalized['population_n'] = normalized['population'] / normalized['population'].max()
normalized['population_n2'] = normalized['population'] / normalized['population'].sum()
normalized = normalized.sort_values(by='population', ascending=False)
normalized
```
