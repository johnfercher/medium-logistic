## Imports

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None 
```

## Load Edges

```python
edges = pd.read_csv('edges.csv')

trains = edges.query('type == "train"')
trains['duration'] = (trains['distance'] / trains['speed']) + trains['delay']
trains['price'] = trains['distance'] * trains['price_rate']
trains['worth'] = trains['duration'] / trains['price']

waters = edges.query('type == "water"')
waters['duration'] = (waters['distance'] / waters['speed']) + waters['delay']
waters['price'] = waters['distance'] * waters['price_rate']
waters['worth'] = waters['duration'] / waters['price']

airs = edges.query('type == "air"')
airs['duration'] = (airs['distance'] / airs['speed']) + airs['delay']
airs['price'] = airs['distance'] * airs['price_rate']
airs['worth'] = airs['duration'] / airs['price']

roads = edges.query('type == "road"')
roads['duration'] = (roads['distance'] / roads['speed']) + roads['delay']
roads['price'] = roads['distance'] * roads['price_rate']
roads['worth'] = roads['duration'] / roads['price']

computedEdges = pd.concat([trains, waters, airs, roads])
computedEdges = computedEdges.sort_values(by="worth", ascending=False)
computedEdges
```

## Save Data

```python
computedEdges.to_csv('edges_computed.csv', index=False)
```
